```python
from sklearn import datasets, neighbors
(easy_x, easy_y) = datasets.make_classification(n_samples=400, n_features = 2, n_informative = 2,
                             n_redundant = 0, n_repeated = 0, n_clusters_per_class=1, class_sep=2)
scatter(easy_x[:,0], easy_x[:,1], c = easy_y, cmap = 'cool')
```

```python
def KNN(dataSet, sample, labels):
    # Find Distances Between Sample and all of the dataSet
    distance = []
    x = 0
    minIndex = 0
    # Calculate all distances
    for i in dataSet:
        x = np.sqrt((sample[0]-i[0])**2 + (sample[1] - i[1])**2)
        distance = np.append(distance, x)
    # Begin finding the nearest neighbors
    minIndex = np.argmin(distance, axis = 0)
    #
    return labels[minIndex]
```

```python
easy_y = np.append(easy_y, KNN(easy_x, [1,1], easy_y))
easy_x = np.append(easy_x, [[1,1]], axis = 0)
```
